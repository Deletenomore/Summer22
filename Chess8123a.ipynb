```python
!pip install chess
import chess, chess.pgn, chess.svg
import io
import matplotlib.pyplot as plt
from IPython.display import SVG, display
```

```python
# Utilities
BLACK = 1
WHITE = 0
def MoveNumber(move, color):
  return move*2 if color else move*2-1

def DisplayLocation(depth, color, chessGame, displayMove=False):
  pgn = io.StringIO(chessGame)
  game = chess.pgn.read_game(pgn)
  board = game.board()
  iter = 0
  for move in game.mainline_moves():
    moveSan = board.san(move)
    board.push(move)
    iter = iter + 1
    if iter >=MoveNumber(depth, color) : break
  if displayMove:
    print(moveSan)
  display(SVG(chess.svg.board(board=board,size=400))  )

def DisplayDepth(depth, chessGame, depthStart=1):
  pgn = io.StringIO(chessGame)
  game = chess.pgn.read_game(pgn)
  board = game.board()
  iter = 0
  for move in game.mainline_moves():
    moveSan = board.san(move)
    board.push(move)
    iter = iter + 1
    if (iter % 2) == 0 and iter >= MoveNumber(depthStart, WHITE):
      print (f'\n{iter // 2}. {moveWhite} {moveSan}')
      display(SVG(chess.svg.board(board=board,size=400))  )
    else:
      moveWhite= moveSan
    if iter >=MoveNumber(depth, BLACK) : break
```
